<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
import matplotlib.pyplot as plt

fig, abx = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True)
a1 = abx[0].plot(x, tan,color = 'b')
a2 = abx[1].plot(x, sin, color = 'r')
a3 = abx[2].plot(x, cos, color = 'g')
abx[0].set_title("Grafik Tan", loc='center')
abx[1].set_title("Grafik Sin", loc='center')
abx[2].set_title("Grafik Cos", loc='center')


line_labels=['TanX','SinX','CosX']

fig.legend([a1, a2, a3],     # The line objects
           labels=line_labels,   # The labels for each line
           loc="upper center",   # Position of legend
           )


plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.0, 0.0, 1, 1])
axes2 = fig.add_axes([-0.1, 0.85, 0.35, 0.3])
axes3 = fig.add_axes([ 0.9,-0.175, 0.35, 0.3])
axes4 = fig.add_axes([-0.1, -0.175, 0.35, 0.3])
axes5 = fig.add_axes([0.9, 0.85, 0.35, 0.3])



axes.plot(x, y, 'b')
axes.set_title('Grafik CosX')
axes2.plot(x, y, 'b')
axes2.plot(x, y5, 'y')
axes2.set_title('Grafik CosX^5')
axes3.plot(x, y, 'b')
axes3.plot(x, y3, 'g')
axes3.set_title('Grafik CosX^3')
axes4.plot(x, y, 'b')
axes4.plot(x, y2, 'r')
axes4.set_title('Grafik CosX^2')
axes5.plot(x, y, 'b')
axes5.plot(x, y4, 'black')
axes5.set_title('Grafik CosX^4')
```

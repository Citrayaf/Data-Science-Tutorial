Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note : warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

figur, ff = plt.subplots()
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2

ff.plot(x,cos, color='r',label='Cos x')
ff.plot(x,cos3, linestyle='dashed',label='3 Cos x',color='g')
ff.plot(x, sin, linestyle='-.',label='Sin x', color='b')
ff.plot(x, sin2,marker='x', linestyle='None',label= '3 Sin x', color='y')

ff.set_xlabel('X-Axis Label')
ff.set_ylabel('Y-Axis Label')
ff.set_title('Grafik Cos Sin')
ff.legend(loc="center")

plt.show()

```
